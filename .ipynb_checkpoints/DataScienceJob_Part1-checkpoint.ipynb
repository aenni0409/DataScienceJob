```python
import requests
import requests_cache
import lxml.html as lx
import re
import json 
import urllib
import time
import numpy as np
import pandas as pd
import plotly.plotly as py
import pandas as pd
import matplotlib.pyplot as plt
import plotly.offline as offline
import plotly.graph_objs as go
import nltk
from nltk.corpus import stopwords
import json
from operator import itemgetter
plt.style.use('ggplot')
```

```python
def fetch_page(url):
    response = requests.get(url)
    # Throw an error if status isn't 200 OK.
    response.raise_for_status()
    return response
```

```python
# Be polite to cache the request and prevent from requesting overwhelmly.
requests_cache.install_cache('demo_cache')
def parse_jobkey(search_title='Data Scientist',page=None):
    jt = '+'.join(str.split(search_title))
    url = 'https://www.indeed.com/jobs?q={0}&l=United+States&start={1}'
    jobs = fetch_page(url.format(jt,page))
    html_string = lx.fromstring(jobs.text)
    path_key = html_string.xpath("//td[@id='resultsCol']//h2[@class='jobtitle']/a/@href")
    path_key_id = [re.split('\?|\&|\=',i) for i in path_key]
    path_key_id = [(i[2],i[6]) for i in path_key_id if 'jk' in i and 'vjs' in i]
    return(path_key_id)
```

```python
def parse_jobdetail(keys, search_title='Data Scientist'):
    data = dict({'search_title':[],'job_title':[],'location':[],'Description':[]})
    count = 0
    for jk,vjs in keys:
        count += 1
        if count%20 == 0:
            print('running on detail...%s/%s'%(str(count),str(len(keys))))
            #setting 'pause' to prevent the request crash
            time.sleep(np.random.choice(20,1))
        url_jd = 'https://www.indeed.com/viewjob?jk={0}&from=serp&vjs={1}'
        try:
            list_page_jd = fetch_page(url_jd.format(jk,vjs))
            html_jd = lx.fromstring(list_page_jd.text)
            data['search_title'].append(search_title)
            data['job_title'].append(html_jd.xpath("//h3/text()")[0])
            location_class = 'jobsearch-InlineCompanyRating icl-u-xs-mt--xs  icl-u-xs-mb--md'
            description_class = 'jobsearch-JobComponent-description icl-u-xs-mt--md'
            data['location'].append(html_jd.xpath("//div[@class='{0}']/div/text()".format(location_class)))
            data['Description'].append(html_jd.xpath("//div[@class='{0}']//text()".format(description_class)))
        except:
            next
    return data
```

```python
def data_collecting(data_approx=2000,save_key=False):
    r = round(data_approx/(5*8))
    save_keys = []
    data = []
    title_list = ['Data Scientist','Data Analyst','Data Engineer',
                  'Business Analyst','Machine learning Engineer']
    for search_title in title_list:
        time.sleep(np.random.choice(5,1))
        count = 0
        print(search_title)
        keys = []
        for i in range(r):
            count += 1
            if count%5 == 0:
                print('running on keys...%s/%s'%(str(count),str(r)))
                time.sleep(np.random.choice(5,1))
            keys += parse_jobkey(search_title,page=i*10)
        data.append(parse_jobdetail(keys,search_title))
        save_keys += keys
    if save_key :
        with open('keys.txt',w) as f:
            f.write(save_keys)
    return(pd.concat([pd.DataFrame(data[i]) for i in range(len(data))],axis=0),)
```

```python
#Extract Programming Language Items from Context
def programming_list():
    with open('programming_language_list.json') as f:
        programming = json.load(f)
    programming = [i['item']["name"] for i in programming['itemListElement']]
    #xml language
    with open('ml_list.json') as f:
        ml = json.load(f)
    ml = [i['name'] for i in ml]

    #BI Tool
    BI = []
    with open('BI_list.txt') as f:
        for r in f:
            BI+=r.strip().split('[ ]')
    BI = [i.split('[')[1].split(']')[0] for i in BI]
    return [i.lower() for i in programming+ml+BI]
#programming Language item all
prog = programming_list()
prog = list(set(prog)-set(['clean','processing','make','plus','focus','code','k']))
```

```python
def description_cleaning(data,search_title):
    title = search_title
    a = [''.join(i) for i in data[data['search_title']==title]['Description']]

    #convert the word to lowercase and splitted.
    a = [s.lower().split() for s in a if s]
    # Convert the list of lists into a flat list
    a = [sublist for l in a for sublist in l]

    #Remove punctuation and numbers from words
    a = [''.join(c for c in w if c.isalpha()) for w in a]
    #Remove stopwords from words
    filtered_sentence = [w for w in a if not w in stop_words and w.isalpha()] 
    return filtered_sentence
```

```python
#calculate the frequency

title_list = ['Data Scientist','Data Analyst','Data Engineer',
              'Business Analyst','Machine learning Engineer']
freq_filter_all = dict()
freq_bi_new_all = dict()

for title in title_list:
    print(title)
    filtered_sentence = description_cleaning(data,title)
    text = nltk.Text(filtered_sentence)
    freq = nltk.FreqDist(text)
    bigrams = nltk.bigrams(text)# Get Bigrams from text
    freq_bi = dict(nltk.FreqDist(bigrams)) # Calculate Frequency Distribution for Bigrams
    keyjoin = [' '.join(i) for i in freq_bi.keys()]
    keyjoinnew = list()
    #flip the mirror to one
    freq_bi_new = dict()
    print('counting for the bi-gram...')
    for i in freq_bi.keys():
        if ' '.join((i[1],i[0])) in keyjoin:
            if ' '.join((i[1],i[0])) not in keyjoinnew:
                freq_bi_new[i] = freq_bi[i] + freq_bi[(i[1],i[0])]
                keyjoinnew+=[' '.join((i[1],i[0]))]
    freq_bi_new = { ' '.join(k): v for k, v in freq_bi_new.items() }
    programm_text = nltk.FreqDist(np.array(filtered_sentence)[[i in prog for i in filtered_sentence]])
    freq_filter = sorted(programm_text.most_common(40),key=itemgetter(1))
    removebi = ['years experience','experience working','equal opportunity',
                'veteran status','team members','color religion','new york',
                'status disability','minimum qualifications','gender identity',
                'race_color','work closely','help us','preferred qualifications',
                'care health', 'job description', 'employment without',
                'experience experience','religion sex','join team','marital status']
    for i in removebi:
        if i in freq_bi_new.keys():
            del freq_bi_new[i]
    freq_bi_new = pd.DataFrame.from_dict(freq_bi_new,orient='index')
    freq_bi_new = freq_bi_new.rename(columns={0:'counts'})
    freq_filter_all[title]=freq_filter
    freq_bi_new_all[title]=freq_bi_new.sort_values(by = ['counts'],ascending=False)['counts'][0:100]
```

```python
import plotly.plotly as py
import plotly.graph_objs as go
pair = ['Data Scientist','Data Analyst']
x_text=map(lambda x:x[0], freq_filter_all[pair[0]])
x_data=pd.DataFrame(list(map(lambda x:x[1], freq_filter_all[pair[0]])),
                      index=list(x_text))
x_data=x_data.rename(columns={0:pair[0]})
x_data[pair[0]]=x_data[pair[0]]/sum(x_data[pair[0]])

y_text=map(lambda x:x[0], freq_filter_all[pair[1]])
y_data=pd.DataFrame(list(map(lambda x:x[1], freq_filter_all[pair[1]])),index=list(y_text))
y_data=y_data.rename(columns={0:pair[1]})
y_data[pair[1]]=y_data[pair[1]]/sum(y_data[pair[1]])

mergedata = x_data.merge(y_data,left_index=True,right_index=True,how='outer')
mergedata = mergedata.fillna(0)
mergedata = mergedata[np.sum(mergedata,1)>.015]




trace0 = go.Scatter(
    x=mergedata[pair[0]],
    y=mergedata[pair[1]],
    mode='markers',
    text=mergedata.index,
    marker=dict(
        color='rgb(252,141,89)',
        size=(mergedata[pair[0]]+mergedata[pair[1]])*400,
    )
)
data = [trace0]

layout = go.Layout(
    title='Relative Frequency of Programming Skills Comparison',
    xaxis=dict(
        title=pair[0],
        gridcolor='rgb(255, 255, 255)',
        type='log',
        zerolinewidth=1,
        ticklen=5,
        gridwidth=2,
    ),
    yaxis=dict(
        title=pair[1],
        gridcolor='rgb(255, 255, 255)',
        type='log',
        zerolinewidth=1,
        ticklen=5,
        gridwidth=2,
    ),
    paper_bgcolor='rgb(243, 243, 243)',
    plot_bgcolor='rgb(243, 243, 243)',
)



fig = dict( data=data, layout=layout )
offline.init_notebook_mode()
offline.plot(fig,filename='DAvsDS.html')


#offline.iplot(data, filename='bubblechart-size')
```
